```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english')
```

```python
matrix = cv.fit_transform(df[df['target']==1]['text'])
freqs = zip(cv.get_feature_names(), matrix.sum(axis=0).tolist()[0])    
print("Top 20 words used for disaster.")
print(sorted(freqs, key=lambda x: -x[1])[:20])
```

```python
matrix = cv.fit_transform(df[df['target']==0]['text'])
freqs = zip(cv.get_feature_names(), matrix.sum(axis=0).tolist()[0])    
print("Top 20 words used for non-disaster tweets.")
print(sorted(freqs, key=lambda x: -x[1])[:20])
```

```python
train = df.iloc[:7613]
test = df.iloc[7613:]
```

```python
X_train = train['text']
y_train = train['target']

X_test = test['text']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
```

```python
print(y_train.shape)
print(X_train.shape)
```

```python
pipe = Pipeline([('tfidf', TfidfVectorizer()),('svc', LinearSVC()),])
```

```python
pipe.fit(X_train, y_train)  
```

```python
predictions = pipe.predict(X_test).astype(int)
predictions
```

```python
predictions.shape
```

```python
df_for_id = pd.read_csv('../input/nlp-getting-started/test.csv')
```

```python
submission = pd.DataFrame({'target':predictions})
submission['id'] = df_for_id['id']
submission = submission.iloc[:, ::-1]
submission['target']= submission['target'].astype('int')
submission.to_csv('submission.csv',index = False)
submission
```
